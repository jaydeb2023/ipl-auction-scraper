```python
url = "https://www.iplt20.com/auction/2022"
print(url)
```

```python
response = requests.get(url, headers=headers)
print(response.status_code)  # Check if the request was successful
```

```python
soup = BeautifulSoup(response.text, 'lxml')
print(soup)
```

```python
table=soup.find("table",class_="ih-td-tab w-100 auction-tbl")
print(table)
```

```python
soup = BeautifulSoup(response.text, 'lxml')
table = soup.find("table", class_="ih-td-tab w-100 auction-tbl")
if table:
    print("Table found!")
```

```python
    title = table.find_all("th")
    header = [i.text.strip() for i in title]  # Extract and clean the header text
    print("Headers:", header)
```

```python
df = pd.DataFrame(columns=header)
```

```python
rows = table.find_all("tr")
```

```python
for row in rows[1:]:
        cols = row.find_all("td")
        if len(cols) > 0:
            # Extract and clean the text for each column
            row_data = [col.text.strip() for col in cols]
            # Append the row data to the DataFrame
            df.loc[len(df)] = row_data
```
